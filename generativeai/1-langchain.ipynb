### Streaming

Streaming means the AI sends its response piece by piece as it generates it, instead of waiting until the entire response is finished.

Without streaming:

You send a prompt.
The model generates the full answer.
You receive everything at once.

With streaming:

You send a prompt.
The model starts generating.
You receive chunks of text immediately.
The answer appears progressively, similar to someone typing.

```python
llm.stream("Write a 500 word essat on Open AI")
```

```python
for chunk in llm.stream("Write a 500 word essay on Iran- US War"):
    print(chunk.text, end = "", flush = True)
```

### Batch Streaming

Batching

Multiple requests are grouped together and processed in one model call.

Goal: improve throughput and reduce cost.

```python
model.batch(
    ['Why do India have a neural policy?',
     "Supreme Court of India",
     "The Library of Congress in US?"],
    config = {
        'max_concurrency' : 5    # Limit to 5 parallel calls
    }
)
```
