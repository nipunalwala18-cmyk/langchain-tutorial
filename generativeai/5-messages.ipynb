### Messages

Messages are the fundamental unit of context for models in LangChain. They represent the input and output of models, carrying both the content and metadata needed to represent the state of a conversation when interacting with an LLM. Messages are objects that contain:

Role – Identifies the message type (e.g. system, user)

Content – Represents the actual content of the message (like text, images, audio, documents, etc.)

Metadata – Optional fields such as response information, message IDs, and token usage data

LangChain provides a standard message type that works across all model providers, ensuring consistent behavior regardless of the model being called.

```python
import os
from langchain.chat_models import init_chat_model

os.environ['GROQ_API_KEY'] = os.getenv("GROQ_API_KEY")

llm = init_chat_model("groq:qwen/qwen3-32b")
```

#### Text Prompts

Text Prompts are strings - which are ideal for straightforward generation tasks where you dont need to retain conversation history

-> You have a single, standalone request

-> You don’t need conversation history

-> You want minimal code complexity

```python
llm.invoke("What is langchain")
```

#### Message Prompts

Alternatively, you can pass in a list of messages to the model by providing a list of message objects.

##### Message types:

1. System message – Tells the model how to behave and provide context for interactions

2. Human message – Represents user input and interactions with the model

3. AI message – Responses generated by the model, including text content, tool calls, and metadata

4. Tool message – Represents the outputs of tool calls

##### When to Use What
Text Prompt

Single question
, No chat history
, Simple LLM calls

Message Prompt

Chatbots
, Agents
, Memory/history
, Tool calling
, Multi-turn conversations

## Example - 1

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage('You are a data science expert'),
    HumanMessage('Explain CERN technology')
]

response = llm.invoke(messages)
print(response.content)
print("\n")
print(response)
print("\n")
print(messages)
```

## Example - 2

```python
sysmessage = SystemMessage("You are a helpful coding assistant")

messages = [
    sysmessage,
    HumanMessage("Explain me about Python Django Framework")
]

response = llm.invoke(messages)
print(response.content)
```

```python
### Detailed Info to LLM with system message

sysmessage = SystemMessage("""You are a senior developer with mastery in Python and expertise with frameworks. 
                           Always provide code examples and explain with reasoning
                           Be concise but thorough in your explanations""")


messages = [
    sysmessage,
    HumanMessage("How do I create a REST API")
]

print(llm.invoke(messages).content)
```

#### Message Metadata


```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    HumanMessage(content='Hello! What is the day today?', 
                name = "alice",     #Optional: identify different users
                id = "msg123")      #Optional: Unique identifier for tracing
]

response = llm.invoke(messages)
print(response.content)
```
